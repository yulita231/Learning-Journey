```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

df_yt=pd.read_csv('/content/youtube-top-100-songs-2025.csv')
print(df_yt.head())
print(df_yt.info())
print(df_yt.describe())
```

```python
print(df_yt.isnull().sum())
```

```python
df_yt=pd.DataFrame([df_yt.duplicated().sum()], columns = ['Duplicated'])
print(df_yt)
```

```python
df_yt=pd.read_csv('/content/youtube-top-100-songs-2025.csv')
median_tags = df_yt['tags'].mode()[0]
df_yt['tags'] = df_yt['tags'].fillna(median_tags)
print(df_yt.isnull().sum())
```

```python
print("Nilai unik dri kolom categoris")
df_yt["categories"].nunique()

# Tampilkan semua nilai unik beserta hitungannya
print(df_yt['categories'].value_counts())
```

```python
plt.figure(figsize=(15,5)) #menentukan ukuran keseluruhan gambar / figure yg akan dibuat
#sns.boxplot       ----> perintah membuat kotak dengan sumbu x dan Y
#data=df_yt         ----> sebagai sumber data yg akan di tampilkan
#palette = "pastel" ----> perintah menentukan warna dengan warna yg pastel

sns.boxplot(x="channel", y="duration" , data=df_yt, palette = "pastel")
plt.title("Song Duration by Channel") #memberikan judul gambar
plt.xlabel("Channel") #nama untuk sumbu X
plt.ylabel("Duration (sec)") #nama untuk sumbu Y
plt.xticks(rotation = 90) #memutar label x agar mudah di baca

plt.show #menampilkan gambar
```
